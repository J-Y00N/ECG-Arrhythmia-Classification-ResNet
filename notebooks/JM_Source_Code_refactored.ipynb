```python
from pathlib import Path

import pandas as pd

from ecg_classification.data import build_dataset_bundle, class_distribution

PROJECT_ROOT = Path.cwd()
TRAIN_CSV = PROJECT_ROOT / "data" / "mitbih" / "mitbih_train.csv"
TEST_CSV = PROJECT_ROOT / "data" / "mitbih" / "mitbih_test.csv"

bundle = build_dataset_bundle(
    train_csv=TRAIN_CSV,
    test_csv=TEST_CSV,
    validation_size=0.10,
    random_state=42,
)

print("Train distribution:", class_distribution(bundle.y_train))
print("Validation distribution:", class_distribution(bundle.y_valid))
print("Test distribution:", class_distribution(bundle.y_test))
```

```python
import matplotlib.pyplot as plt
import numpy as np

from ecg_classification.constants import LABEL_TO_SYMBOL

x_axis = np.arange(bundle.X_train.shape[1])

fig, ax = plt.subplots(figsize=(14, 6))
for label in sorted(LABEL_TO_SYMBOL):
    index = np.where(bundle.y_train == label)[0][0]
    ax.plot(x_axis, bundle.X_train[index], label=LABEL_TO_SYMBOL[label])

ax.set_title("Representative heartbeat by class")
ax.set_xlabel("Sample index")
ax.set_ylabel("Amplitude")
ax.legend()
plt.show()
```
